## Introduction

One of the most fundamental and extensively researched issues in computer science is sorting. It entails placing a group of things in a predetermined order, such as ascending, descending, alphabetical, or numerical, such as a collection of words, numbers, or records. Sorting is helpful for many things, including making it simpler to search, merge, filter, and do analyses on data. Various fields, including database management, information retrieval, cryptography, and computer graphics, all have uses for sorting. An array or list's elements can have their order changed using a sorting algorithm. A comparison operator that we use on the elements determines the new order. The placement of the items in the data structure is determined by the comparison operator. [1]

A collection of items can be sorted in a variety of ways, and each method has benefits and drawbacks of its own. Sorting algorithms are techniques that can efficiently and accurately perform sorting for various types of data and criteria. Sorting algorithms can be categorized using a variety of criteria, including:

- Complexity: How much time and space the algorithm needs to sort a specific input size. We can use Big O notation or "asymptotic analysis" to compare the effectiveness of various sorting methods. This informs us of the number of operations an algorithm requires to completely sort a really large data collection. [2] As an illustration, an algorithm with O(n2) time complexity sorts n items in around n2 steps, where n is the input size. The input's initial order, such as whether it is already sorted, reverse sorted, or randomly jumbled, can also affect how complex the problem is.

 - Stability: The capacity to maintain the relative order of equal elements in the input is known as stability. [3] For instance, a stable sorting algorithm would maintain the order of two things with the same value but different qualities, such as names or colors, but an unstable sorting algorithm may alter the order of the items.
 
- In-place: The capability of sorting the input with no or little extra space required. For instance, a non-in-place sorting algorithm will use another array or data structure to hold the sorted elements while an in-place sorting algorithm will sort the input array by swapping or shifting its elements within the array itself.

- Comparator: The algorithm that chooses how to compare two input elements. A comparator function, for instance, could compare the values of two numbers, the lengths of two words, or the keys to two records. Additionally, a comparator function can be modified to accommodate various requirements or preferences. [4]

- Comparison-based: Refers to a property where the order of the elements is determined by comparison or by using some other data, such as their values or positions. A non-comparison sorting algorithm will sort the input by using some other information about the items, such as their digits or bits, whereas a comparison-based sorting algorithm will sort the input by continually comparing pairs of items and determining which one should come first. [5]

## The 5 Sorting Algorithms we will use:

### Bubble Sort

1. Bubble Sort: A straightforward and reliable comparison-based sort which repeatedly swaps out-of-order nearby elements until the array is sorted [6]. Although bubble sort is simple to use and comprehend, it is incredibly slow and ineffective for inputs that are large or nearly sorted. In both the worst and average cases, the time complexity of bubble sort is O(n2), where n is the size of the input array. When the array is already sorted or nearly so, the time complexity is O(n) in the best case scenario. Since swapping only requires a small amount of extra space, bubble sort has an O(1) space complexity.

### Selection Sort

2. Selection Sort: A straightforward in-place comparison-based sort that consistently locates the smallest item in the array's unsorted portion and places it there [7]. Although selection sort is also simple to use and understand, it is also very slow and ineffective for inputs that are large or nearly sorted. Selection sort always has an O(n2) time complexity, where n is the size of the input array. Selection sort only requires a constant amount of extra space for swapping, hence its space complexity is O(1).

### Insertion Sort

3. Insertion Sort: This straightforward, dependable, and in-place comparison-based sort constructs the sorted array one item at a time by inserting each item into the appropriate location [8]. For small or nearly sorted inputs, insertion sort is more effective than bubble sort and selection sort, but it is still slow and inefficient for large or randomly shuffled inputs. In both the worst and average cases, the time complexity of insertion sort is O(n2), where n is the size of the input array. When the array is already sorted or nearly so, the time complexity is O(n) in the best case scenario. Since insertion sort only requires a constant amount of extra space for shifting, its space complexity is O(1).

### Quick Sort

4. Quick Sort: An efficient and in-place comparison-based sort that partitions the array into two parts based on a pivot item, recursively sorts each part, and combines them back together [9]. One of the most popular sorting algorithms is quick sort since it works quickly and effectively on the majority of inputs. For some inputs, such as already sorted or reverse sorted arrays, quick sort can, however, be extremely slow and inefficient. Quick sort typically has an O(n log n) time complexity, where n is the size of the input array. When the array is already sorted or nearly so, the worst case time complexity is O(n2). Due to the recursive calls, quick sort has an O(log n) average space complexity and an O(n) worst case.

### Radix Sort

5. Radix Sort: A non-comparison sort that sorts the array by grouping items based on their digits or bits, starting from the least significant one [10]. Radix sort is very fast and efficient for sorting integers or strings with fixed length or range, as it does not depend on comparing items. However, because it needs more passes and space, radix sort can be very slow and ineffective when sorting items with variable length or range. Radix sort has an O(dn) time complexity, where d denotes the number of digits or bits and n is the size of the input array. Radix sort has an O(d+n) space complexity, where d is the number of digits or bits and n is the length of the input array.

## Bubble Sort

The fundamental sorting technique known as bubble sort involves verifying and exchanging nearby entries in a list repeatedly until they are in the proper order. Because of the way it resembles bubbles rising to the surface of water, it is called Bubble Sort.



### Example of Bubble Sort

Let's say we have a list of numbers that we want to sort in ascending order:

[4, 1, 1, 3, 6, 2]

We start with the first element and evaluate it against the subsequent element. In this case, we switch the two elements if the first element is larger. Other than that, we leave them alone. Up until the end of the list, this is done for every group of adjacent elements. This is the algorithm's initial iteration.

The biggest part of the list will be at the end after the first pass:

[1, 4, 1, 3, 2, 6]

The final piece is left alone since it is already in its proper position, and we repeat the process for the other elements. This is the algorithm's second pass.

The second-largest element will be at the second-to-last spot on the list following the second pass:

[1, 1, 3, 2, 4, 6]

With the exception of the final two, which are already in their proper locations, we repeat the process for the remaining pieces. This is the algorithm's third pass.

The third-largest element will be at the third-last spot on the list following the third pass:

[1, 1, 2, 3, 4, 6]

The final three pieces are left out because they are already in their proper placements. We then repeat the process for the remaining elements. This is the algorithm's fourth pass.

The fourth-largest element will be at the fourth-last spot on the list following the fourth pass:

[1, 1, 2, 3, 4, 6]

We see that the list is already sorted and that no further comparisons or swaps are required. Here, the algorithm can be stopped.

Here is the final, sorted list:

[1, 1, 2, 3, 4, 6]


### Python Code for Implementation of Bubble Sort

```python
def bubbleSort(arr):
    # get the length of the array
    n = len(arr)

    # loop through each element of the array
    for i in range(n-1):

        # keep track of swapping
        swapped = False

        # loop to compare adjacent elements
        for j in range(0, n-i-1):

            # compare two adjacent elements
            # change > to < to sort in descending order
            if arr[j] > arr[j+1]:

                # swapping occurs if elements
                # are not in the intended order
                arr[j], arr[j+1] = arr[j+1], arr[j]

                # set swapped to True
                swapped = True

        # no swapping means the array is already sorted
        # so no need for further comparison
        if not swapped:
            break

# test the algorithm with an example array
arr = [4, 1, 1, 3, 6, 2]

# call the bubbleSort function and pass the array
bubbleSort(arr)

# print the sorted array
print("Sorted array is:")
for i in range(len(arr)):
    print(arr[i], end=" ")

# References: [11] [12] 
```

## Selection Sort

The simple sorting method known as "Selection Sort" is repeatedly identifying the least (or largest) element from the array's unsorted section and putting it at the start (or end) of the section that has been sorted. It is called Selection Sort because it selects the smallest (or largest) element on each iteration.



## Example of Selection Sort

We have an array that we wish to arrange ascendingly:

[4, 1, 1, 3, 6, 2]

Starting with the first element, we discover the least element across the entire array and swap it with it. The first element is sorted in this manner.

[1, 4, 1, 3, 6, 2]

The minimum element is then selected from the array's remaining unsorted portion (apart from the first element) and is switched with the second element. The second element is sorted in this manner.

[1, 1, 4, 3, 6, 2]

This procedure is repeated for the remainder of the array's unsorted components (aside from the first two elements), and the third member is substituted. The third element is arranged in this manner.

[1, 1, 2, 3, 6, 4]

This procedure is carried through until the array's final element is reached. When that happens, the array as a whole is sorted.

[1, 1, 2, 3, 4, 6]

The array is now sorted and no more selection or swapping is needed. We can stop the algorithm here.

### Python Code for Implementation of Selection Sort

```python
def selectionSort(arr):
    # get the length of the array
    n = len(arr)

    # loop through each element of the array
    for i in range(n):

        # find the minimum element in the remaining unsorted part
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j

        # swap the minimum element with the current element
        arr[i], arr[min_index] = arr[min_index], arr[i]

# test the algorithm with an example array
arr = [4, 1, 1, 3, 6, 2]

# call the selectionSort function and pass the array
selectionSort(arr)

# print the sorted array
print("Sorted array is:")
for i in range(len(arr)):
    print(arr[i], end=" ")


# Reference: [27]
```

## Insertion Sort

Insertion sort is a straightforward sorting method, which places each element in the appropriate place within a subarray that has already been sorted. It resembles how we arrange playing cards in our hands to sort them.  [33]



### Example of Insertion Sort

On our test array of [4, 1, 1, 3, 6, 2], let's use insertion sort.

- As a sorted subarray, element 4 is our first choice.
- If the second point 1 and the first point 4 don't match, we compare them and exchange them. In this manner, we obtain a subarray [1, 4] with two sorted entries.
- The identical process of comparison with the elements of the sorted subarray and placement in the appropriate position is carried out by each element of the array.
- The second element 4 and the third element 1 are switched after the comparison.Therefore, since the first element of 1 and 1 are equal, there is nothing to be done. The subarray after sorting is now [1, 1, 4].
- We contrast and swap the third and fourth elements, respectively. After comparing 3 to the subsequent element 1, we come to an end because 3 is greater than 1. The subarray is now [1, 1, 3, 4] after sorting.
- We do not trade when comparing the fifth element to the fourth element because 6 is greater than 4, thus we do not. The subarray that has been sorted is now [1, 1, 3, 4, 6].
- We contrast and swap the sixth element 2 with the fifth element 6. In order to swap them, we compare 2 to the fourth element of 4. Therefore, we swap the elements of 2 and 3 after comparing them.The second element of 1 is being compared to 2, and since 2 is greater than 1, we stop. Now the subarray is sorted, and it is [1, 1, 2, 3, 4, 6].

### Python Code for Implementation of Insertion Sort

```python
def insertionSort(arr):
    # if array has one or zero elements, it is already sorted
    if len(arr) <= 1:
        return
    # loop from index 1 to n-1
    for i in range(1, len(arr)):
        # store the current element as key
        key = arr[i]
        # initialize j as i-1
        j = i - 1
        # loop while j is not negative and key is smaller than arr[j]
        while j >=0 and key < arr[j]:
            # move arr[j] to arr[j+1]
            arr[j+1] = arr[j]
            # decrement j by one
            j -= 1
        # place key at arr[j+1]
        arr[j+1] = key

# test the code
arr = [4,1,1,3,6,2]
insertionSort(arr)
print(arr)

Reference: [32]
```

### Time Complexity of Insertion Sort

The time complexity of insertion sort depends on how sorted the array is initially.

- The best case scenario is when the array is already sorted or nearly sorted. In this case, insertion sort only needs to compare each element with its left neighbour once and perform no swaps. Therefore, the time complexity is O(n), where n is the number of elements in the array.
- The worst case scenario is when the array is in reverse order. In this case, insertion sort needs to compare each element with all the elements in the sorted subarray and perform n(n-1)/2 swaps. Therefore, the time complexity is O(n^2).
- The average case scenario is when the array is randomly ordered. In this case, insertion sort needs to compare each element with about half of the elements in the sorted subarray and perform n(n-1)/4 swaps. Therefore, the time complexity is also O(n^2). [45]

### Space Complexity of Insertion Sort

The space complexity of insertion sort is O(1), because it only uses a constant amount of extra space to store some variables. It does not create any new arrays or data structures to sort the array.

## Quick Sort

Quick sort is a fast and efficient sorting algorithm that works by dividing and conquering the array into smaller subarrays. It is based on the idea of choosing a pivot element and partitioning the array around it. [37]



###  Example of Quick Sort

Let's see how quick sort works on the our array [4, 1, 1, 3, 6, 2].

- We start by choosing the rightmost element 2 as the pivot element.
- We rearrange the array so that elements smaller than the pivot are on the left and elements larger than the pivot are on the right. We also swap the pivot with the first element larger than it. This way, we have a partitioned array [1, 1, 2, 3, 6, 4].
- We repeat this process for the left subarray [1, 1] and the right subarray [3, 6, 4], choosing new pivots and partitioning them. The left subarray is already sorted, so we do nothing. The right subarray is partitioned as [3, 4, 6] with 6 as the pivot.
- We repeat this process for the left subarray [3] and the right subarray [], choosing new pivots and partitioning them. Both subarrays are already sorted, so we do nothing.

### Python Code for the Implementation of Quick Sort

```python
def quickSort(arr, low, high):
    # if array has one or zero elements, it is already sorted
    if low < high:
        # partition the array around a pivot
        pi = partition(arr, low, high)
        # sort the left subarray
        quickSort(arr, low, pi - 1)
        # sort the right subarray
        quickSort(arr, pi + 1, high)

def partition(arr, low, high):
    # choose the rightmost element as pivot
    pivot = arr[high]
    # initialize i as low - 1
    i = low - 1
    # loop from low to high - 1
    for j in range(low, high):
        # if element is smaller than pivot
        if arr[j] < pivot:
            # increment i by one
            i += 1
            # swap arr[i] and arr[j]
            arr[i], arr[j] = arr[j], arr[i]
    # swap arr[i + 1] and pivot
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    # return i + 1 as the partition index
    return i + 1

# test the code
arr = [4,1,1,3,6,2]
n = len(arr)
quickSort(arr, 0, n - 1)
print(arr)

# Reference: [36]
```

### Quick sort diagram in Python

```python
# Import the matplotlib.pyplot module
import matplotlib.pyplot as plt

# Import the random module
import random

# Define a function to swap two elements in a list
def swap(lst, i, j):
    lst[i], lst[j] = lst[j], lst[i]

# Define a function to partition a list around a pivot
def partition(lst, low, high):
    # Declare the global counter variable
    global counter
    # Choose the last element as the pivot
    pivot = lst[high]
    # Set the index of smaller element to low - 1
    i = low - 1
    # Loop through the elements from low to high - 1
    for j in range(low, high):
        # If the current element is smaller than or equal to the pivot
        if lst[j] <= pivot:
            # Increment the index of smaller element
            i += 1
            # Swap the current element with the smaller element
            swap(lst, i, j)
            # Update the global counter of swaps
            counter += 1
            # Plot the list after swapping on the grid of plots
            ax = plt.subplot(4, 4, counter) # Create a subplot on the grid based on the counter
            ax.bar(range(len(lst)), lst) # Plot the list as a bar chart
            ax.set_xticks(range(len(lst))) # Label the x axis with indices
            yticks = range(min(lst), max(lst) + 1) # Create a range of numbers from min to max of list values
            ax.set_yticks(yticks) # Label the y axis with numbers from min to max
            ax.set_xlabel("Position") # Set the label of the x axis to "Position"
            ax.set_ylabel("Number") # Set the label of the y axis to "Number"
            ax.set_title(f"Swapping {lst[i]} and {lst[j]}") # Set the title of the plot
            
    # Swap the pivot with the element at i + 1
    swap(lst, i + 1, high)
    # Update the global counter of swaps
    counter += 1
    
# Plot the list after swapping on the grid of plots
    ax = plt.subplot(4, 4, counter) # Create a subplot on the grid based on the counter
    ax.bar(range(len(lst)), lst) # Plot the list as a bar chart
    ax.set_xticks(range(len(lst))) # Label the x axis with indices
    yticks = range(min(lst), max(lst) + 1) # Create a range of numbers from min to max of list values
    ax.set_yticks(yticks) # Label the y axis with numbers from min to max
    ax.set_xlabel("Position") # Set the label of the x axis to "Position"
    ax.set_ylabel("Number") # Set the label of the y axis to "Number"
    ax.set_title(f"Swapping {lst[i + 1]} and {lst[high]}") # Set the title of the plot
    
    # Return the index of the pivot
    return i + 1

# Define a function to perform quicksort on a list and create a grid of plots
def quicksort(lst, low, high):
    # If low is less than high
    if low < high:
        # Partition the list and get the index of the pivot
        pi = partition(lst, low, high)
        # Recursively sort the left and right sublists and update the grid of plots
        quicksort(lst, low, pi - 1)
        quicksort(lst, pi + 1, high)

# Create a list of 6 elements from [4, 1, 1, 3, 6, 2]
lst = [4, 1, 1, 3, 6, 2]

# Initialize a global counter of swaps to keep track of plot positions on the grid
counter = 0

# Create a figure with a grid of 4 by 4 subplots 
plt.figure(figsize=(16, 16))

# Plot the list before sorting on the first subplot 
ax = plt.subplot(4, 4, 1)
ax.bar(range(len(lst)), lst)
ax.set_xticks(range(len(lst))) 
ax.set_yticks(range(min(lst), max(lst) + 1)) 
ax.set_xlabel("Position") 
ax.set_ylabel("Number") 
ax.set_title("Before sorting")

# Perform quicksort on the list and update the grid of plots 
quicksort(lst, 0, len(lst) - 1)

# Adjust the spacing between subplots 
plt.tight_layout()

# Show all subplots
plt.show()
```

### Time Complexity of Quick Sort

The time complexity of quick sort depends on how balanced the partitions are.

- The best case scenario is when the pivot element divides the array into two equal or nearly equal subarrays. In this case, quick sort performs n log n comparisons and swaps. Therefore, the time complexity is O(n log n), where n is the number of elements in the array.
- The worst case scenario is when the pivot element is either the smallest or the largest element in the array. In this case, quick sort performs n^2 comparisons and swaps. Therefore, the time complexity is O(n^2).
- The average case scenario is when the pivot element divides the array into two subarrays of varying sizes. In this case, quick sort performs about n log n comparisons and swaps. Therefore, the time complexity is also O(n log n). [46]

### Space Complexity of Quick Sort

The space complexity of quick sort is O(log n), because it uses a recursive call stack to sort the subarrays. It does not create any new arrays or data structures to sort the array.


### Advantages and Disadvantages of using Quick Sort

Advantages:

- It is one of the fastest sorting algorithms with an average time complexity of O(n log n).
- It is a divide-and-conquer algorithm that breaks down the original problem into smaller subproblems that are easier to solve.
- It is an in-place algorithm that does not require any extra space.

Disadvantages:

- It has a quadratic time complexity of O(n^2) in the worst case, which makes it slow for large or unsorted arrays.
- It is not stable, meaning it does not preserve the relative order of equal elements.
- It is sensitive to the choice of pivot element, which can affect its performance and balance. [38]

## Radix Sort

Radix sort is a sorting algorithm that sorts the elements by grouping them by their individual digits at the same position. It uses another stable sorting algorithm, such as counting sort, to sort the elements according to each digit. [39]



### Example of Radix Sort

Let's see how radix sort works on our array [4, 1, 1, 3, 6, 2].

- We start by finding the maximum element in the array, which is 6. It has 1 digit, so we only need to sort the array once based on the unit place digit.
- We use counting sort to sort the elements based on their unit place digits. We create 10 buckets for digits from 0 to 9 and count the frequency of each digit in the array. Then we place the elements in their corresponding buckets according to their unit place digits. The result is [1, 1, 2, 3, 4, 6].

Reference: [40]

### Implementation of Radix Sort in Python

```python
def radixSort(arr):
    # find the maximum element in the array
    max_element = max(arr)
    # find the number of digits in the maximum element
    digits = len(str(max_element))
    # initialize a variable to store the place value
    place = 1
    # loop from 0 to digits - 1
    for i in range(digits):
        # use counting sort to sort the elements based on place value
        arr = countingSort(arr, place)
        # multiply the place value by 10
        place *= 10
    # return the sorted array
    return arr

def countingSort(arr, place):
    # find the length of the array
    n = len(arr)
    # initialize an output array of size n
    output = [0] * n
    # initialize a count array of size 10 (for digits from 0 to 9)
    count = [0] * 10
    # loop through the array and count the frequency of each digit at place value
    for i in range(n):
        # find the digit at place value
        digit = (arr[i] // place) % 10
        # increment the count of that digit by one
        count[digit] += 1
    # calculate the cumulative sum of the count array
    for i in range(1, 10):
        count[i] += count[i - 1]
    # loop through the array from right to left and place the elements in output array according to their count
    for i in range(n - 1, -1, -1):
        # find the digit at place value
        digit = (arr[i] // place) % 10
        # find the index of that digit in output array using count array
        index = count[digit] - 1
        # place the element at that index
        output[index] = arr[i]
        # decrement the count of that digit by one
        count[digit] -= 1
    # return the output array
    return output

# test the code
arr = [4,1,1,3,6,2]
arr = radixSort(arr)
print(arr)

# Reference: [41]
```

### Time Complexity of Radix Sort

The time complexity of radix sort depends on how many digits are there in each element and how many elements are there in the array.

- The best case scenario is when all elements have only one digit. In this case, radix sort only needs to perform one pass of counting sort, which takes O(n + k) time, where n is the number of elements and k is the range of digits (10 in this case). Therefore, the time complexity is O(n + 10), which is O(n).
- The worst case scenario is when all elements have d digits, where d is large. In this case, radix sort needs to perform d passes of counting sort, each taking O(n + k) time. Therefore, the time complexity is O(d * (n + 10)), which is O(dn).
- The average case scenario is when the elements have varying number of digits. In this case, radix sort needs to perform d passes of counting sort, where d is the maximum number of digits among all elements. Therefore, the time complexity is also O(dn).

### Space Complexity of Radix Sort

The space complexity of radix sort is O(n + k), because it uses an output array of size n and a count array of size k (10 in this case) to store the intermediate results. It does not create any new arrays or data structures to sort the array. [42]

## Implementation and Benchmarking

This section describes the implementation and benchmarking of five different sorting algorithms in Python on various input sizes. The sorting algorithms used are:

- Bubble Sort: A basic sorting algorithm that compares adjacent elements and swaps them if they are out of order.
- Insertion Sort: A basic sorting algorithm that places each element in its correct position in a sorted subarray.
- Radix Sort: A sorting algorithm that does not compare elements but sorts them by their digits or letters.
- Quick Sort: An efficient sorting algorithm that divides the array around a pivot element and sorts the subarrays on either side of the pivot recursively.
- Selection Sort: A basic sorting algorithm that finds the minimum element in the array and moves it to the beginning.

The programme consists of a main Python script called benchmarking.py that imports the sorting algorithm functions from separate .py files, creates random arrays of integers of varying input sizes, and measures the running time of each sorting algorithm. The project also uses several modules such as random, time, pandas, matplotlib.pyplot, seaborn, and tabulate to generate, record, store, format, and display the results. 

The main function benchmarking takes in three arguments:

- algorithms: A dictionary containing the names of the sorting algorithms and the corresponding functions imported from other files.
- input_sizes: A list containing the different input sizes to be tested, ranging from 10 to 10000.
- runs: An integer indicating the number of times each sorting algorithm is applied to each input size.

The function then creates four empty lists to store the elapsed times, run numbers, sort functions, and input sizes for each trial. It then loops through each sorting algorithm provided in the dictionary, each input size provided in the list, and each run number provided by the range function. For each combination of these parameters, it does the following steps:

- It generates a new random array of integers using the random.randint function with 0 and 10000 as the lower and upper bounds respectively.
- It records the current time using the time.time function before calling the sorting algorithm function on the random array. It also passes values for low and high indices if required by some algorithms such as quick sort.
- It records the current time again using the time.time function after the sorting algorithm function returns. It then computes the elapsed time by subtracting the start time from the end time.
- It adds the elapsed time, run number, sort function name, and input size to their respective lists. [52]

After completing all the iterations, the programme creates a pandas dataframe with four columns: Elapsed Time, Run Number, Sorting Algorithm, and Input Size. It then groups the dataframe by Sorting Algorithm and Input Size and calculates the average elapsed time for each combination. A pivot table is created with Sorting Algorithm as columns and Input Size as rows and Average Elapsed Time as values. [54] The tabulate function is used to format the pivot table as a plain text table with borders and alignment. The programme then prints the table to the console with a caption indicating the number of runs. Seaborn's lineplot function is then used to display the pivot table with markers and labels. We set the title, xlabel, and ylabel of the plot and it is shown on the screen. [55]

The application also exports the dataframe of individual trial results and the pivot table of average results to two separate csv files for reference. [51]

The Benchmarking Code uses the following sources and tools:

- Python 3.9.7
- Random module for creating random integers
- Time module for measuring time in seconds
- Pandas module for creating and manipulating dataframes
- Matplotlib.pyplot module for plotting graphs
- Seaborn module for enhancing graphs

## Python Benchmarking Code

```python
import random
import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tabulate

# import sorting algorithm functions from separate .py files
import bubble_sort
import insertion_sort
import radix_sort
import quick_sort
import selection_sort


# function to generate random arrays of integers
def random_array(n):
    return [random.randint(0, 10000) for _ in range(n)]

def benchmarking(algorithms, input_sizes, runs):
    
    # create empty lists to store results
    elapsed_times = []
    run_numbers = []
    sort_functions = []
    input_sizes_list = []

    # iterate through each sorting algorithm provided
    for sort_name, sort_func in algorithms.items():

        # iterate through each input size provided
        for size in input_sizes:

            # repeat the sorting process for a number of runs
            for run in range(runs):

                # generate a random array of integers for each run
                arr = random_array(size)

                # create a copy of the array for the sorting function to operate on
                arr_copy = arr.copy()

                # time the sorting function applied to the random array copy
                start_time = time.time()
                # pass 0 and len(arr_copy) - 1 as low and high arguments for quickSort function
                sorted_arr = sort_func(arr_copy, 0, len(arr_copy) - 1) if sort_name == "Quick Sort" else sort_func(arr_copy)
                end_time = time.time()

                # calculate elapsed time and append results to lists
                elapsed_time = (end_time - start_time) * 1000
                elapsed_times.append(elapsed_time)
                run_numbers.append(run+1)
                sort_functions.append(sort_name)
                input_sizes_list.append(size)

                # ensure the sorting function has actually sorted the array
                # assert sorted_arr == sorted(arr)

    # create pandas dataframe with results of each individual trial
    df = pd.DataFrame({
        "Elapsed Time (ms)": elapsed_times,
        "Run Number": run_numbers,
        "Sorting Algorithm": sort_functions,
        "Input Size (n)": input_sizes_list
    })

   
    # group by sorting algorithm and input size and calculate average elapsed time for each combination
    df_grouped = df.groupby(["Sorting Algorithm", "Input Size (n)"]).mean().reset_index()

    # format elapsed time to 2 decimal places and convert to float
    df_grouped["Elapsed Time (ms)"] = df_grouped["Elapsed Time (ms)"].apply(lambda x: float("{:.2f}".format(x)))

    # format input sizes using scientific notation
    df_grouped["Input Size (n)"] = df_grouped["Input Size (n)"].apply(lambda x: "{:.0f}".format(x))

    # add column for number of runs conducted
    df_grouped["Number of Runs"] = runs

    # sort table by input size in ascending order
    df_grouped = df_grouped.sort_values(by=["Input Size (n)"])


    # create a pivot table with sorting algorithms as columns and input sizes as rows and average elapsed times as values
    df_pivot = df_grouped.pivot(index="Input Size (n)", columns="Sorting Algorithm", values="Elapsed Time (ms)")

    # use the tabulate function to format the pivot table as a plain text table with borders and alignment
    print(tabulate.tabulate(df_pivot, headers="keys", tablefmt="fancy_grid", showindex=True))
    
    # print a caption for the table
    print("Sorting Algorithms - Average of 10 runs")

    # plot the pivot table using seaborn lineplot with markers and labels 
    sns.lineplot(data=df_pivot, markers=True)
    plt.title("Sorting Algorithms - Average Times")
    plt.xlabel("Input Size (n)")
    plt.ylabel("Elapsed Time (s)")
    plt.show()

# define a dictionary of sorting algorithms to be tested
algorithms = {
    "Bubble Sort": bubble_sort.bubble_sort,
    "Insertion Sort": insertion_sort.insertion_sort,
    "Radix Sort": radix_sort.radixSort,
    "Quick Sort": quick_sort.quickSort,
    "Selection Sort": selection_sort.selectionSort,
}

# define input sizes to be tested
input_sizes = [100, 250, 500, 750, 1000, 1250, 2500, 3750, 5000, 6250, 7500, 8750, 10000]


# define number of runs for each combination of sorting algorithm and input size
runs = 10

# call the benchmarking function with the defined algorithms, input sizes, and runs
benchmarking(algorithms, input_sizes, runs)
```
